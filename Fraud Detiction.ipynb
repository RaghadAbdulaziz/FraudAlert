```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
cards_data = pd.read_csv('/content/archive (1).zip')
```

```python
cards_data.head()
```

```python
cards_data.info()
```

```python
cards_data.isnull().sum()
```

**Preprocessing**

```python
cards_data['Class'].value_counts()
```

0 --> legit transaction,
1 --> ilegal transaction,
Data is unbalanced

```python
legit = cards_data[cards_data.Class==0]
fraud = cards_data[cards_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
cards_data.groupby('Class').mean()
```

```python
# select randomly from data, to make it balanced
legit_sample = legit.sample(492)
```

```python
new_dataset= pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# splitting data into features & targets
X= new_dataset.drop(columns='Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting data into train & test data
x_train, x_test, y_train, y_test= train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the model with training data
model.fit(x_train, y_train)
```

Model Evaluation

```python
# accuracy on training data
x_train_prediction= model.predict(x_train)
train_data_accuracy= accuracy_score(x_train_prediction, y_train)
```

```python
print("training data accuracy:" ,train_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy= accuracy_score(x_test_prediction, y_test)
```

```python
print("accuracy score on test data:", test_data_accuracy)
```
